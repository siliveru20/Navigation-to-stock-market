```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
```

```python
# Fetch historical stock data for Tesla (TSLA)
ticker = 'TSLA'
data = yf.download(ticker, start="2010-01-01", end="2023-01-01")

# Display the first few rows of the data
print(data.head())
```

```python
# Select only the 'Close' price
close_prices = data['Close'].values
close_prices = close_prices.reshape(-1, 1)

# Scale the data (normalization between 0 and 1)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(close_prices)

# Define training and testing data split (80% train, 20% test)
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]
```

```python
def create_dataset(data, time_step=60):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

# Prepare the training dataset
X_train, y_train = create_dataset(train_data)

# Reshape the data to fit the LSTM model (samples, time steps, features)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
```

```python
# Build the LSTM model
model = Sequential()

# Add LSTM layers with dropout
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))

model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))

# Add a Dense layer to predict the output
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=64)
```

```python
# Prepare the test dataset
X_test, y_test = create_dataset(test_data)

# Reshape the test data
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Make predictions on the test data
predictions = model.predict(X_test)

# Reverse the scaling to get the original price
predictions = scaler.inverse_transform(predictions)
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
def predict_future_prices(model, last_data, future_days=30, time_step=60):
    # Create an empty list to store the future predictions
    future_predictions = []

    # Start with the last 'time_step' days
    current_input = last_data[-time_step:].reshape(1, time_step, 1)

    for _ in range(future_days):
        # Predict the next day's price
        predicted_price = model.predict(current_input)

        # Append the predicted price to the future_predictions list
        future_predictions.append(predicted_price[0, 0])

        # Update the input to include this prediction (shift the window)
        predicted_price_reshaped = predicted_price.reshape(1, 1, 1)  # Reshape the predicted price to 3D
        current_input = np.append(current_input[:, 1:, :], predicted_price_reshaped, axis=1)

    # Return the future predictions, scaled back to the original price range
    return scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))
```

```python
# Number of days to predict
future_days = 30

# Ensure your last 60 days of data (or similar) are available for prediction
last_60_days = test_data[-60:]  # Adjust the number of days as per your input size

# Generate the future predictions
predicted_future_prices = predict_future_prices(model, last_60_days, future_days=future_days)
```

```python
# Define 'ticker' if it's not already defined
ticker = "Your Stock Ticker"  # Replace with the actual stock ticker

# Plot actual vs predicted for the test period
plt.figure(figsize=(12, 6))

# Plot the actual prices
plt.plot(np.arange(len(y_test_rescaled)), y_test_rescaled, label='Actual Prices')

# Plot the predicted future prices (starts after the actual price data)
plt.plot(np.arange(len(y_test_rescaled), len(y_test_rescaled) + future_days),
         predicted_future_prices, label=f'Predicted Prices for next {future_days} days', linestyle='--')

# Add titles and labels
plt.title(f'{ticker} Stock Price Prediction')
plt.xlabel('Days')
plt.ylabel('Price (USD)')
plt.legend()

# Show the plot
plt.show()
```
